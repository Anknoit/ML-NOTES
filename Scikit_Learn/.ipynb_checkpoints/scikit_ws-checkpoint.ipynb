## KNN Classification - Model

K Nearest Neighbour - Classifying as well as regression modelm

```python
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
# Learning from example - Here we are using KNN classification in car data sets to classify car's [class(label) - unacc, acc,good,vgood] while training our model using features - 'buying', 'maintenance', 'safety'
```

```python
#Loading dataset
data = pd.read_csv('car.data')
data.head()
```

```python
# Deciding and Assigning Features(X) and Label(y) we gonna use 
X = data[['buying', 'maint', 'safety']].values
y = data[['class']]
# print(X,y)
```

```python
# Converting character datavalues of X and y to numerical 2D-array to feed into ML model
Le = LabelEncoder()
# X conversion - Using labelEncoder
for i in range(len(X[0])):
    X[:, i] = Le.fit_transform(X[:, i])
# print(X)

# y conversion - using smple mapping function map()
label_mapping = {
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}
y['class'] = y['class'].map(label_mapping)
y = np.array(y)
# print(y)
```

```python
# Creating the model
knn = neighbors.KNeighborsClassifier(n_neighbors=25, weights='uniform')
# n_neighbors = K
# weights = ?

#Separating data into 2 categories Training data and Testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#Training the model
knn.fit(X_train, y_train) #requires 2 parameters X,y. It feeds the data into the model to train

#Predictions
prediction = knn.predict(X_test)

#Accuracy Score
Accuracy = metrics.accuracy_score(y_test, prediction)

print("Predictions" ,prediction)
print("Accuracy:", Accuracy*100,"%")

# Testin individual data using indexing
indi_test_position = 1400
print('Actual Value: ', y[indi_test_position])
print('Predicted Value: ', knn.predict(X)[indi_test_position])
```

```python
plt.title("PREDICTIONS")
plt.ylabel("Predicted Value - 3=Good, 0= Unacceptable")
plt.plot(prediction)
```
